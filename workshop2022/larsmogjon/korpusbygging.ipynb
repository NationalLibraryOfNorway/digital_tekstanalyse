```python
import dhlab as dh
import pandas as pd
```

```python
df = pd.read_csv('../data/partipressen_normalized.csv', index_col=0)
```

```python
df
```

```python
parti_dct = dict()
partier = df.parti.unique()

for parti in partier:
    parti_dct[parti] = df.loc[df.parti == parti, 'normalisert']
```

```python
def corpus_looper(s):
    c = dh.text.corpus.EmptyCorpus()
    
    for x in s:
        delkorpus = dh.Corpus(doctype='digavis', title=x, from_year=1945, to_year=1970, limit=100)
        
        c.add(delkorpus)
        
    return c
        
        
```

```python
df_dct = dict()

for parti in partier:
    df_dct[parti] = corpus_looper(parti_dct[parti])
```
