### 1.1. N-gram for ordparadigmene

For en introduksjon til n-gram, se eksempelfil om [N-gram og galakser](https://nbviewer.jupyter.org/github/DH-LAB-NB/DHLAB/blob/master/4_N-gram_og_galakser.ipynb) eller [NB N-gram.](https://www.nb.no/sp_tjenester/beta/ngram_1/)

Lenke til [Helene Uri](https://www.aftenposten.no/kultur/i/awkxj2/Helene-Uri-Dette-er-en-reklametekst-for-Norges-stiligste-nettside)

Her har vi en funksjon som viser ngram for ordene i word_paradigm, `paradigme_ngram`.

`y=` definerer hvilken av ordgruppene vi får ut. Default er satt til 0.
`z=` her kan vi angi ord utenfor paradigmet som vi også ønsker å undersøke.

```python
paradigme_ngram = lambda x, y =0, z='': nb.nb_ngram(','.join(nb.word_paradigm(x)[y][1]+[z])).plot(figsize=(10,5),lw=(3))
```

```python
paradigme_ngram("fisker", y=1, z='')
```
